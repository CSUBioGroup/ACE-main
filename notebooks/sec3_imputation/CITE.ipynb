## loading data

```python
root_dir = '../../data/CITE'
dataset_path = os.path.join(root_dir, 'openproblems_bmmc_cite_phase2_mod2.censor_dataset.output_')

par = {
    'input_train_mod1': f'{dataset_path}train_mod1.h5ad',
    'input_train_mod2': f'{dataset_path}train_mod2.h5ad',
    'input_train_sol': f'{dataset_path}train_sol.h5ad',
    'input_test_sol': f'{dataset_path}test_sol.h5ad',
    'input_test_mod1': f'{dataset_path}test_mod1.h5ad',
    'input_test_mod2': f'{dataset_path}test_mod2.h5ad',
}

print('Reading `h5ad` files...')
input_train_mod1 = sc.read_h5ad(par['input_train_mod1'])
input_train_mod2 = sc.read_h5ad(par['input_train_mod2'])
input_train_sol = sc.read_h5ad(par['input_train_sol'])
input_test_mod1 = sc.read_h5ad(par['input_test_mod1'])
input_test_mod2 = sc.read_h5ad(par['input_test_mod2'])

print("Unscrambling training cells...")
ord = input_train_sol.X.tocsr().indices
if "pairing_ix" in input_train_sol.uns:
    assert np.all(ord == np.argsort(input_train_sol.uns["pairing_ix"]))
input_train_mod2 = input_train_mod2[ord, :].copy()
input_train_mod2.obs_names = input_train_mod1.obs_names

input_test_sol = sc.read_h5ad(par['input_test_sol'])
ord = input_test_sol.X.tocsr().indices
if "pairing_ix" in input_test_sol.uns:
    assert np.all(ord == np.argsort(input_test_sol.uns["pairing_ix"]))
input_test_mod2 = input_test_mod2[ord, :].copy()
input_test_mod2.obs_names = input_test_mod1.obs_names    

# select hvg for evaluation
sc.pp.highly_variable_genes(input_train_mod2, n_top_genes=5000)
hvg_idx = np.where(input_train_mod2.var.highly_variable)[0]
```

```python
log_dir = '../../outputs/CITE-imp'
```

## RNA->ADT

```python
os.makedirs(join(log_dir, 'RNA2ADT'), exist_ok=True)
gex = input_train_mod2.copy()
other = input_train_mod1.copy()
gex_test = input_test_mod2.copy()
other_test = input_test_mod1.copy()

#### preprocessing
if not os.path.exists(join(log_dir, 'RNA2ADT/gex_train_input.npy')):
    ### GEX preprocessing
    gex_all = sc.concat([gex, gex_test])
    lsi_former = lsiTransformer(192, drop_first=True, use_highly_variable=False, log=False, norm=False, 
                                z_score=True, tfidf=False, svd=True, use_counts=False
    )
    gex_all_lsi_df = lsi_former.fit_transform(gex_all)
    gex_all_dimred = HARMONY(gex_all_lsi_df, gex_all.obs.batch.to_list(), use_gpu=True)
    gex.obsm['dimred_be'], gex_test.obsm['dimred_be'] = gex_all_dimred[:gex.shape[0]], gex_all_dimred[gex.shape[0]:]

    del gex_all, gex_all_lsi_df, gex_all_dimred
    gc.collect()

    # ADT processing
    adt_lsi_former = ADTransformer(other.shape[1], drop_first=False, svd=False, log=False, 
                                     norm=False, z_score=True, use_counts=False)
    adt_df = adt_lsi_former.fit_transform(other)
    other.obsm['X_be'] = HARMONY(adt_df, other.obs.batch.to_list(), use_gpu=True)
    other_test.obsm['X_be'] = np.zeros(other_test.shape)

    del adt_df

    np.save(join(log_dir, 'RNA2ADT/gex_train_input.npy'), gex.obsm['dimred_be'])
    np.save(join(log_dir, 'RNA2ADT/gex_test_input.npy'),  gex_test.obsm['dimred_be'])
    np.save(join(log_dir, 'RNA2ADT/other_train_input.npy'), other.obsm['X_be'])

else:
    gex.obsm['dimred_be']      = np.load(join(log_dir, 'RNA2ADT/gex_train_input.npy'))
    gex_test.obsm['dimred_be'] = np.load(join(log_dir, 'RNA2ADT/gex_test_input.npy'))
    other.obsm['X_be']     = np.load(join(log_dir, 'RNA2ADT/other_train_input.npy'))   
    other_test.obsm['X_be']= np.zeros(other_test.shape)
```

```python
n_parts = 2
modBatch_dict = {
    'rna': [gex, gex_test],
    'adt': [other, None]
}

useReps_dict = {
    'rna': 'dimred_be',
    'adt': 'X_be'
}
```

```python
# building stage 1
T = 0.1
model = ACE(
    modBatch_dict=modBatch_dict,  
    useReps_dict=useReps_dict,  
    batch_key='batch', 
    layers_dims = {'rna': [1024, 512], 'adt':[512, 2048]}, # consistent across all experiments
    dropout_rates = {'rna':[0.2, 0.2], 'adt':[0.2, 0.2]},  # consistent across all experiments
    T=math.log(1./T), T_learnable=False, log_dir=join(log_dir, 'RNA2ADT'),
    n_latent1=40, n_latent2=256, seed=1234, num_workers=0
)

model.stage1_fit(
    batch_size=512,
    epochs=100,
    lr=2e-4, 
    log_step=10, 
)
```

```python
# stage1 inference
ad_integ = model.stage1_infer(
    modBatch_dict, useReps_dict, output_key='stage1_emb', 
    specify_mods_perBatch=[['rna'], ['rna']]
)
ad_integ
```

```python
# impute
imputed_data = model.impute(modBatch_dict, output_key1='stage1_emb', knn=20, verbose=True)
Y_hat = imputed_data['adt'][1]
```

## ADT -> RNA

```python
os.makedirs(join(log_dir, 'ADT2RNA'), exist_ok=True)
gex = input_train_mod2.copy()
other = input_train_mod1.copy()
gex_test = input_test_mod2.copy()
other_test = input_test_mod1.copy()

#### preprocessing
if not os.path.exists(join(log_dir, 'ADT2RNA/gex_train_input.npy')):
    ### GEX preprocessing
    lsi_former = lsiTransformer(192, drop_first=True, use_highly_variable=False, log=False, norm=False, 
                                z_score=True, tfidf=False, svd=True, use_counts=False
    )
    gex_df = lsi_former.fit_transform(gex)
    gex.obsm['dimred_be'] = HARMONY(gex_df, gex.obs.batch.to_list(), use_gpu=True)
    gex_test.obsm['dimred_be'] = np.zeros((gex_test.shape[0], 192))

    del gex_df
    gc.collect()

    # ADT processing
    other_all = sc.concat([other, other_test])
    adt_lsi_former = ADTransformer(other.shape[1], drop_first=False, svd=False, log=False, 
                                     norm=False, z_score=True, use_counts=False)
    adt_all_df = adt_lsi_former.fit_transform(other_all)
    adt_all_arr = HARMONY(adt_all_df, other_all.obs.batch.to_list(), use_gpu=True)
    other.obsm['X_be'], other_test.obsm['X_be'] = adt_all_arr[:other.shape[0]], adt_all_arr[other.shape[0]:]

    del adt_all_df, adt_all_arr, other_all
    gc.collect()

    np.save(join(log_dir, 'ADT2RNA/gex_train_input.npy'), gex.obsm['dimred_be'])
    np.save(join(log_dir, 'ADT2RNA/other_train_input.npy'), other.obsm['X_be'])
    np.save(join(log_dir, 'ADT2RNA/other_test_input.npy'), other_test.obsm['X_be'])
else:
    gex.obsm['dimred_be']      = np.load(join(log_dir, 'ADT2RNA/gex_train_input.npy'))
    gex_test.obsm['dimred_be'] = np.zeros((gex_test.shape[0], 192))
    other.obsm['X_be']    = np.load(join(log_dir, 'ADT2RNA/other_train_input.npy'))
    other_test.obsm['X_be'] = np.load(join(log_dir, 'ADT2RNA/other_test_input.npy'))
```

```python
n_parts = 2
modBatch_dict = {
    'rna': [gex, None],
    'adt': [other, other_test]
}

useReps_dict = {
    'rna': 'dimred_be',
    'adt': 'X_be'
}
```

```python
# building stage 1
T = 0.1
model = ACE(
    modBatch_dict=modBatch_dict,  
    useReps_dict=useReps_dict,  
    batch_key='batch', 
    layers_dims = {'rna': [1024, 512], 'adt':[512, 2048]}, # consistent across all experiments
    dropout_rates = {'rna':[0.2, 0.2], 'adt':[0.2, 0.2]},  # consistent across all experiments
    T=math.log(1./T), T_learnable=False, log_dir=join(log_dir, 'ADT2RNA'),
    n_latent1=40, n_latent2=256, seed=1234, num_workers=0
)

model.stage1_fit(
    batch_size=512,
    epochs=100,
    lr=2e-4, 
    log_step=10, 
)
```

```python
# stage1 inference
ad_integ = model.stage1_infer(
    modBatch_dict, useReps_dict, output_key='stage1_emb', 
    specify_mods_perBatch=[['rna'], ['adt']]
)
ad_integ
```

```python
# impute
imputed_data = model.impute(modBatch_dict, output_key1='stage1_emb', knn=20, verbose=True)
X_hat = imputed_data['rna'][1]
```
