### loading data

```python
# root_dir = '/home/yanxh/gitrepo/multi-omics-matching/neurips2021_multimodal_topmethods-main'
data_dir = "../../data/DOGMA"

print('Reading `mtx` files...')
cite_rna_mat = sps.csr_matrix(sio.mmread(join(data_dir, 'CITE-seq/rna_mat_count_full.mtx')).T)
cite_adt_mat = sps.csr_matrix(sio.mmread(join(data_dir, 'CITE-seq/adt_mat_count.mtx')).T)
asap_atac_mat= sps.csr_matrix(sio.mmread(join(data_dir, 'ASAP-seq/atac_mat_count.mtx')).T)
asap_adt_mat = sps.csr_matrix(sio.mmread(join(data_dir, 'ASAP-seq/adt_mat_count.mtx')).T)
# cite_rna_names = pd.read_csv(join(data_dir, 'CITE-seq/hvg_names.csv'))['VariableFeatures(CITE_rna)'].to_numpy()
cite_rna_names = pd.read_csv(join(data_dir, 'CITE-seq/rna_names_full.csv')).iloc[:, 1].values
cite_adt_names = pd.read_csv(join(data_dir, 'CITE-seq/adt_names.csv'))['VariableFeatures(CITE_ASAP_adt)'].to_numpy()
asap_atac_names = pd.read_csv(join(data_dir, 'ASAP-seq/hvp_names.csv'))['VariableFeatures(ASAP_atac)'].to_numpy()
asap_adt_names = pd.read_csv(join(data_dir, 'ASAP-seq/adt_names.csv'))['VariableFeatures(CITE_ASAP_adt)'].to_numpy()

cite_cell_names = pd.read_csv(join(data_dir, 'CITE-seq/cell_names.csv'))['x'].to_numpy()
asap_cell_names = pd.read_csv(join(data_dir, 'ASAP-seq/cell_names.csv'))['x'].to_numpy()

cite_meta_data = pd.read_csv(join(data_dir, 'CITE-seq/metadata.csv'), index_col=0)
asap_meta_data = pd.read_csv(join(data_dir, 'ASAP-seq/metadata.csv'), index_col=0)

cite_meta_data['celltypes_from_scVAEIT'] = cite_meta_data['celltypes'].to_numpy()
asap_meta_data['celltypes_from_scVAEIT'] = asap_meta_data['celltypes'].to_numpy()

# this is the original annotations: NK, T cell, B cell, Myeloid
type2dict = {'NK':'NK', 'CD4 T': 'T cell', 'CD8 T': 'T cell', 'B': 'B cell', 'Mono': 'Myeloid', 'DC': 'Myeloid'}
cite_meta_data['cell_type'] = [type2dict[_] for _ in cite_meta_data.celltypes]
asap_meta_data['cell_type'] = [type2dict[_] for _ in asap_meta_data.celltypes]

print('Creating `h5ad` files...')
ad_cite_rna = sc.AnnData(cite_rna_mat, obs=cite_meta_data)
ad_cite_adt = sc.AnnData(cite_adt_mat, obs=cite_meta_data)
ad_asap_atac = sc.AnnData(asap_atac_mat, obs=asap_meta_data)
ad_asap_adt  = sc.AnnData(asap_adt_mat,  obs=asap_meta_data)
ad_cite_rna.var_names = cite_rna_names
ad_cite_adt.var_names = cite_adt_names
ad_asap_atac.var_names= asap_atac_names
ad_asap_adt.var_names = asap_adt_names
```

### preprocessing

```python
ad_cite_rna.obsm['dimred'] = sps.csr_matrix(sio.mmread(join(data_dir, 'CITE-seq/rna_harmonypca.mtx')))
ad_cite_adt.obsm['dimred'] = sps.csr_matrix(sio.mmread(join(data_dir, 'CITE-seq/adt_harmonypca.mtx')))
ad_asap_atac.obsm['dimred'] = sps.csr_matrix(sio.mmread(join(data_dir, 'ASAP-seq/atac_harmonypca.mtx')))
ad_asap_adt.obsm['dimred'] = sps.csr_matrix(sio.mmread(join(data_dir, 'ASAP-seq/adt_harmonypca.mtx')))
```

### construct inputs

```python
n_parts = 4

modBatch_dict = {
    'rna': [ad_cite_rna[ad_cite_rna.obs.batch=='control'].copy(), ad_cite_rna[ad_cite_rna.obs.batch=='stim'].copy(), None, None],
    'adt': [
        ad_cite_adt[ad_cite_adt.obs.batch=='control'].copy(), ad_cite_adt[ad_cite_adt.obs.batch=='stim'].copy(), 
        ad_asap_adt[ad_asap_adt.obs.batch=='control'].copy(), ad_asap_adt[ad_asap_adt.obs.batch=='stim'].copy(), 
    ],
    'atac':[None, None, ad_asap_atac[ad_asap_atac.obs.batch=='control'].copy(), ad_asap_atac[ad_asap_atac.obs.batch=='stim'].copy()]
}

useReps_dict = {
    'rna': 'dimred',
    'adt': 'dimred',
    'atac': 'dimred'
}
```

### stage1 training and inference

```python
log_dir = './checkpoints'
```

```python
T = 0.1
model = ACE(
    modBatch_dict=modBatch_dict,  
    useReps_dict=useReps_dict,  
    batch_key='batch', 
    layers_dims = {'rna': [1024, 512], 'adt':[512, 2048], 'atac':[1024, 512]}, # consistent across all experiments
    dropout_rates = {'rna':[0.2, 0.2], 'adt':[0.2, 0.2],  'atac':[0.2, 0.2]},  # consistent across all experiments
    T=math.log(1./T), T_learnable=False, log_dir=log_dir,
    n_latent1=40, n_latent2=256, seed=1234, 
    num_workers=0  # have to set 0 in jupyter notebook, otherwise reporting: https://github.com/jupyter/nbconvert/issues/1066
)

# model.stage1_fit(
#     batch_size=512,
#     epochs=101,
#     lr=2e-4, 
#     log_step=10, 
# )
```

#### load the pretrained the weights

```python
utls.load_model(model.stage1_encoders.encoders, model.mod_names, log_dir, ckpt_id=101, stage_label='stage1')
```

```python
# stage1 inference
ad_integ = model.stage1_infer(
    modBatch_dict, useReps_dict, output_key='stage1_emb', 
    specify_mods_perBatch=[['adt'], ['adt'], ['adt'], ['adt']]  
)
ad_integ
```

## Analyze the batches measured with CITE-seq

```python
ad_cite_rna.obs['louvain'] = adata_manu.obs.loc[ad_cite_rna.obs_names, 'louvain'].to_numpy()
ad_cite_adt.obs['louvain'] = adata_manu.obs.loc[ad_cite_adt.obs_names, 'louvain'].to_numpy()

ad_cite_rna.obsm['X_umap'] = adata_manu[ad_cite_rna.obs_names].obsm['X_umap']
ad_cite_adt.obsm['X_umap'] = adata_manu[ad_cite_adt.obs_names].obsm['X_umap']
```

DE analysis

```python
sc.pp.normalize_total(ad_cite_rna, target_sum=1e4)
sc.pp.log1p(ad_cite_rna)
sc.tl.rank_genes_groups(ad_cite_rna, 'louvain', method='wilcoxon')

sc.pp.normalize_total(ad_cite_adt, target_sum=1e4)
sc.pp.log1p(ad_cite_adt)
sc.tl.rank_genes_groups(ad_cite_adt, 'louvain', method='wilcoxon')
```

```python
uniq_marker_genes = ['CD3E', 'CD3D', 'CD3G', 
                     'CD4', #'TCF7', 'LEF1',  # naive cd4
                     #'IL7R', #'IL32', 'LTB',  # CD4 memory T
                     'CD8A', 'CD8B',   # CD8 
                     'PRF1', 'NKG7', 'GNLY', 'KLRD1',  # nk
                     'MS4A1', 'CD79A', 'RALGPS2',   # B cells
                     'CD14', 'CCDC88A', 'LYZ', 'S100A9',  # cluster 6, CD14 mono,
                     'CCL5', 'CST7',    # 8T memory 
                     'RTKN2', 'FOXP3', 'IL2RA', 'TIGIT',  # Treg
                     'TRDC', 'TRGC1',   # gdT
                     'KLRB1',  # MAIT
                     'ITM2C', 'SERPINF1', 'IL3RA',  # pDC
                     'PRSS57', 'EGFL7', 'GATA2',    # HSPC
                    ]
```

```python
obj = ad_cite_rna
subsample_idx = []
for ci in obj.obs.louvain.unique():
    ci_mask = np.where(obj.obs.louvain==ci)[0]
    if len(ci_mask) >= 100:
        ci_mask = np.random.choice(ci_mask, min(100, int(0.5*len(ci_mask))), replace=False)
        np.random.shuffle(ci_mask)
        subsample_idx.append(ci_mask)
    else:
        subsample_idx.append(ci_mask)

subsample_idx = np.hstack(subsample_idx)
print(subsample_idx.size)

obj = obj[subsample_idx].copy()
sc.pl.heatmap(obj, uniq_marker_genes, groupby='louvain', swap_axes=True, figsize=(20, 8), log=False, cmap='Blues',vmin=0., vmax=2.,
             show=False)
```
